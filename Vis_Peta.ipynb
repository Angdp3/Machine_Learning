```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from adjustText import adjust_text
import pandas as pd

# Tentukan warna untuk setiap cluster
warna_cluster = {
    1: 'red',
    2: 'orange',
    3: 'green'
}

# Buat colormap dari dictionary warna
cmap = colors.ListedColormap(list(warna_cluster.values()))
# Convert 'Cluster' column to numeric
dataku['Cluster'] = pd.to_numeric(dataku['Cluster'])

# Atur ukuran figure untuk landscape A4
fig, ax = plt.subplots(1, 1, figsize=(11.69, 8.27))  # Ukuran A4 landscape dalam inch

# Plot data dengan warna berdasarkan cluster
dataku.plot(
    column='Cluster',
    legend=True,
    ax=ax,
    cmap=cmap,
    categories=[1, 2, 3],
    linewidth=5
)

# Tambahkan label ID
texts = []
for x, y, label in zip(dataku.geometry.centroid.x,
                      dataku.geometry.centroid.y,
                      dataku['NAMOBJ']):
    ax.plot(x, y, marker='o', markersize=3, color='white')
    texts.append(ax.text(x, y, label, fontsize=6, color='black', ha='center', va='top'))
adjust_text(texts,
            force_points=0.3,
            force_text=0.3,
            expand_points=(2, 2),
            expand_text=(1.5, 1.5),
            arrowprops=dict(arrowstyle='-', color='gray', lw=0.5),
            only_move={'points':'', 'text':'xy'},
            )

plt.title('Visualisasi Cluster Desa')
plt.show()
```
